```python
import torch.nn as nn


class MLP(nn.Module):
  def __init__(self):
    super.__init__(MLP)

  def forward(self, x)->nn:
    return nn.Sequential(
        x = nn.Linear(x),
        x = nn.RELU(x),
        x = nn.Linear(x),
        x = nn.RELU(x),
    )
```
